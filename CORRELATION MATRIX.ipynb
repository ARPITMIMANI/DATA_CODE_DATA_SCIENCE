```python
# CORRELATION MATRIX : IT IS SIMPLY A TABLE SHOWING THE CORRELATION COEFFICIENTS BETWEEN THE VARIABLES .

# LOAD THE DATA FROM THE CSV FILE "EXERCISE_DATA.csv" INTO THE DATAFRAME .

import pandas as pd
ww = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
print ( ww.to_string() )
```

```python
# corr() IS USED FOR CALCUATION OF CORRELATION COEFFICIENTS BETWEEN THE VARIABLES .
# CREATE A CORRELATION MATRIX USING PANDAS .

import pandas as pd
ww = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
xx = round ( ww.corr() , 2 )
print ( xx.to_string() )
```

```python
import pandas as pd
ww = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
yy = round ( ww.corr() , 4 )
print ( yy.to_string() )
```

```python
# CREATE A CORRELATION HEATMAP USING SEABORN .

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ww = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
zz = ww.corr()

sns.heatmap ( zz , vmin=-1 , vmax=+1 , center=0 , cmap=sns.diverging_palette ( 50 , 500 , n=500 ) , square=True )
plt.show()
```
