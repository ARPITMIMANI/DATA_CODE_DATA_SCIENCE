```python
'''

INPUT : x
OUTPUT : f(x)

CORRELATION COEFFICIENT : IT ALSO MEASURES THE RELATIONSHIP BETWEEN 2 VARIABLES .
THE VALUE OF CORRELATION COEFFICIENT IS NEVER LESS THAN -1 AND NEVER GREATER THAN +1 .

+1 = PERFECT POSITIVE LINEAR RELATIONSHIP .
0  = NO LINEAR RELATIONSHIP .
-1 = PERFECT NEGATIVE LINEAR RELATIONSHIP .

'''

# EXAMPLE OF PERFECT POSITIVE LINEAR RELATIONSHIP .

import matplotlib.pyplot as plt
import pandas as pd

xx = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
xx.plot ( x="Average_Pulse" , y="Calorie_Burnage" , kind="scatter" )

plt.grid(True)
plt.show()
```

```python
# EXAMPLE OF PERFECT POSITIVE LINEAR RELATIONSHIP .

import matplotlib.pyplot as plt
import pandas as pd

yy = { "absences_in_days" : [40,36,32,28,24,20,16,12,8,4] , "student_scores" : [72,74,76,78,80,82,84,86,88,90] }
zz = pd.DataFrame ( data=yy )

zz.plot ( x="absences_in_days" , y="student_scores" , kind="scatter" )
plt.grid(True)
plt.show()
```
