```python
# PLOT THE VALUES OF "Average_Pulse" AGAINST "Calorie_Burnage" BY USING MATPLOTLIB .

import matplotlib.pyplot as plt
import pandas as pd 

xyz = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )
xyz.plot ( x="Average_Pulse" , y="Calorie_Burnage" , kind="line" )

plt.grid(True)
plt.xlim ( xmin=70 , xmax=130 )
plt.ylim ( ymin=230 , ymax=350 )
plt.show()
```

```python
'''

IN THE ABOVE GRAPH "Average_Pulse" INCREASES BY 10 AND "Calorie_Burnage" INCREASES BY 20 .
SO SLOPE  = 20 / 10 = 2 

MATHEMATICALL FROMULA OF 
SLOPE = ( f(x2) - f(x1) ) / ( x2 - x1 ) = ( y2 - y1 ) / ( x2 - x1 ) 
WHERE ,

y2 = SECOND OBSERVATION OF "Calorie_Burnage" = 260 .
y1 = FIRST OBSERVATION OF "Calorie_Burnage" = 240 .
x2 = SECOND OBSERVATION OF "Average_Pulse" = 90 .
x1 = FIRST OBSERVATION OF "Average_Pulse" = 80 .

slope = ( 260 - 240 ) / ( 90 -80 ) = 20 / 10 = 2 .

'''

# CALCULATE VALUE OF SLOPE USING PYTHON .

def slope ( y1 , y2 , x1 , x2 ) :
    s = ( y2 - y1 ) / ( x2 - x1 )
    return s

t = slope ( 240 , 260 , 80 , 90 )
print(t)
```
