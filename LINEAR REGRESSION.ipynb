```python
# DRAW THE LINE OF LINEAR REGRESSION FOR "Average_Pulse" VS "Calorie_Burnage" .

import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv ( "EXERCISE_DATA.csv" , header=0 , sep="," )

x = df["Average_Pulse"]
y = df["Calorie_Burnage"]

slope , intercept , r , p , std_err = stats.linregress(x,y)

def my_func (x) :
    return slope * x + intercept

mymodel = list ( map ( my_func , x ) )

plt.scatter(x,y)
plt.plot ( x , slope * x + intercept )

plt.xlim ( xmin=60 , xmax=150 )
plt.ylim ( ymin=210 , ymax=350 )

plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")

plt.grid()
plt.show()
```
